```
def Euclidean_distance(x, Y):
    return (np.sum((np.abs(x - Y)) ** 2, axis=1)) ** (1 / 2)
```

```
def GaussianKernel(x,Y, sigma):
  d=Y.shape[1]
  dist=[]
  distances = Euclidean_distance(x,Y)
  for i in range(len(Y)):
    dist_i=(1/((2*(np.pi)**(d/2))*sigma**d))*(np.exp((-1/2)*((distances[i]**2)/(sigma**2))))
    dist.append(dist_i)
  return dist
```

```
def one_hot(c, classes):
    hot = np.zeros(len(classes))
    idx = np.where(classes == c)
    hot [idx]= 1
    return hot
```

```
def confusion_matrix(true_labels, pred_labels):
    n_classes=len(np.unique(true_labels))
    matrix = np.zeros((n_classes, n_classes))
    for (true, pred) in zip(true_labels, pred_labels): 
        matrix[int(true-1)][int(pred-1)] += 1
    return matrix
```

```
def split_dataset(banknote):
    train_set = np.empty((0,5))
    test_set = np.empty((0,5))
    valid_set = np.empty((0,5))
    for i,j in enumerate(banknote):
        r = i % 5
        if( r == 4 ):
            test_set = np.vstack([test_set, j])
            
        elif( r == 3):
            valid_set = np.vstack([valid_set, j])
        else: 
            train_set = np.vstack([train_set, j])
    return (train_set , valid_set , test_set)
```

```
class HardParzen:
    def __init__(self, h):
        self.h = h

    def train(self, train_inputs, train_labels):
        self.label_list = np.unique(train_labels)
        self.train_inputs = train_inputs
        self.train_labels = train_labels
        self.n_classes = len(self.label_list)

    def compute_predictions(self, test_data):
        #test_size = test_data.shape[0]
        #mat = np.ones((test_size,self.n_classes))
        #classes_pred =  np.zeros(test_size)
        classes_pred = []
        for (i , j) in enumerate(test_data):
        #for data_i in test_data:
            dists = Euclidean_distance(j , self.train_inputs)
            ng = []
            #ng = np.array([d for d in range(len(dist)) if dists[d] < float(self.h)])
            for d,dist in enumerate (dists):
                if(dist < float(self.h)):
                    ng.append(d)
            if len(ng) == 0:
                #c = np.zeros( self.n_classes)
                #for ex in self.n_classes:
                    #c.append(draw_rand_label(j, self.label_list) - 1)
                #mat[i, :] = draw_rand_label(j, self.label_list)
                #mat = np.vstack([mat, c])
                classes_pred.append(draw_rand_label(j, self.label_list))
            else:
                labels = 0
                for k in ng:
                    labels = labels+one_hot(self.train_labels[k], self.label_list )
                    
                    #mat[i , self.train_labels[k]-1] += 1
                    #predictions[i] = np.argmax(mat[i, :]) + 1
                label = np.argmax(labels)
                classes_pred.append(self.label_list[label])
                
        return classes_pred
            
```

```
class SoftRBFParzen:
    def __init__(self, sigma):
        self.sigma  = sigma

    def train(self, train_inputs, train_labels):
        self.label_list = np.unique(train_labels)
        self.train_inputs=train_inputs
        self.train_labels=train_labels
        
 
    def compute_predictions(self, test_data):
        #possible_classes= np.unique(self.train_labels)
        classes_pred=[]
        
        for i,test_vector in enumerate(test_data):
            
          distributions = GaussianKernel(test_vector, self.train_inputs , self.sigma)
          dist=0
          for j,d in enumerate(distributions):
            dist = dist + (distributions[j]*one_hot(self.train_labels[j] , self.label_list))
          idx = np.argmax(dist)
          classes_pred.append(self.label_list[idx])
        return classes_pred  

```

```
class ErrorRate:
    def __init__(self, x_train, y_train, x_val, y_val):
        self.x_train = x_train
        self.y_train = y_train
        self.x_val = x_val
        self.y_val = y_val

    def hard_parzen(self, h):
        Hparzen = HardParzen(h)
        Hparzen.train(self.x_train , self.y_train)
        classes_pred = Hparzen.compute_predictions(self.x_val)
        conf_mat = confusion_matrix(self.y_val, classes_pred)
        total_num = np.sum(conf_mat)
        num_correct = np.sum(np.diag(conf_mat))
        return 1.0 - (num_correct / total_num)
        #correct = np.sum(classes_pred == self.y_val)
        #l= len(self.y_val)
        #return (l-correct)/l
     
        

    def soft_parzen(self, sigma):
        Sparzen = SoftRBFParzen(sigma)
        Sparzen.train(self.x_train , self.y_train)
        classes_pred = Sparzen.compute_predictions(self.x_val)
        conf_mat = confusion_matrix(self.y_val, classes_pred)
        total_num = np.sum(conf_mat)
        num_correct = np.sum(np.diagonal(conf_mat))
        #return 1.0 - (num_correct / total_num)
        #s = np.diagonal(conf_mat)
        #num_correct = s.sum()
        return 1.0 - (num_correct / total_num)
        
        
```
